```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
#Para activar los gráficos de notebook
```

```python
P=100 #Tiempo máximo
h=0.001 #Salto de tiempo
r=int(P/h) #Tamaño del arreglo, número de saltos
w=(9.81/10)**.5 #Constante de la fórmula
v=np.zeros(r) #Inicia arreglo V en cero 
ang=np.zeros(r) #Inicia arreglo ang en cero
v[0]=20 #V del arreglo 0 es 20 ya que si no damos valor de v[0] el for no tiene sentido

t = np.linspace (0, r-1, r) #Arreglo utilizado para gráficar el tiempo
```

```python
for i in range(r-1): #Calculos
    v[i+1]=v[i]-(h*w**2)*ang[i]
    ang[i+1]=ang[i]+h*v[i]
```

```python
plt.plot(t/1000, ang, color='k') #Gráfica de angulo contra tiempo
plt.xlabel('Tiempo')
plt.ylabel('Angulo')
```

```python
plt.plot(t/1000,v,color='y') #Más gráficos
plt.xlabel("Tiempo")
plt.ylabel("Velocidad")
```
