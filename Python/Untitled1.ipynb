```python
def recta(a,m=1,b=0,label = 'x'):
    if m<0:
        print('Pendiente')
    z=m*a+b
    plt.plot(a,z,label=label)
```

```python
x=np.linspace(-2,2,20)
recta(x)
recta(x,-2,1,'-x+1')
plt.legend()

plt.grid
```
